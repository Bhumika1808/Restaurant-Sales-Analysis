### Determine the top three most common cuisines in the dataset.

```python
cuisines_counts=df['Cuisines'].value_counts()
```

```python
cuisines_counts
```

```python
all_cui=df['Cuisines'].str.split(',',expand=True).stack()
```

```python
all_cui=all_cui.str.strip()
```

```python
c_counts=all_cui.value_counts()
c_counts
```

```python
top3=c_counts.head(3)
print(f""" The top three most common cuisines in the dataset are:
{top3}""")
```
