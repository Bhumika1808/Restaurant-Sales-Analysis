```python
df=pd.read_csv("Dataset  (1).csv")
df
```

### Identify if there are any restaurant chains present in the dataset.

```python
res_chain=df.groupby("Restaurant Name").size().reset_index(name="Outletcount")
new=res_chain[res_chain["Outletcount"]>1]
res_chains= new.sort_values(by="Outletcount", ascending=False)
```

```python
print(res_chain[["Restaurant Name","Outletcount"]].head(10))
```

```python
chain_counts = df['Restaurant Name'].value_counts()
restaurant_chains = chain_counts[chain_counts > 1]
chain_avg_ratings = df[df['Restaurant Name'].isin(restaurant_chains.index)].groupby('Restaurant Name')['Aggregate rating'].mean().sort_values(ascending=False)
```
