 ### Task: City Analysis
 * Identify the city with the highest number of restaurants in the dataset.
 * Calculate the average rating for restaurants in each city.
 * Determine the city with the highest average rating.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Dataset  (1).csv")
df
```

```python
city_counts = df['City'].value_counts()
```

```python
print("the city with the highest number of restaurants in the dataset is :")
city_counts.head(1)
```

```python
city_avg_ratings = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)
```

```python
print("the average ratings for restaurants in each city are as follows; ")
city_avg_ratings
```

```python
print("the city with the highest average rating.")
city_avg_ratings.head(1)
```
